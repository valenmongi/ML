## **Curacion dataset**

```python
# Eliminamos columnas 'Lat', 'Long' y 'Province/State'

data_1=data.drop(['Lat', 'Long', 'Province/State'], axis=1)
data_1
```

```python
# Pivoteamos la tabla

data_2=pandas.pivot_table(data_1, columns=['Country/Region'], aggfunc=sum).reset_index()
data_2
```

Tendría que ordenar los dias, están cruzados

```python
# Renombramos la columna de las fechas

data_2.rename(columns={'index':'Date'},
             inplace=True)
```

```python
# Configuramos a la columna date en formato fecha

data_2['Date']=pandas.to_datetime(data_2['Date'])
```

```python
# Ponemos en orden las fechas

data_2=data_2.sort_values(by='Date')
```

```python
# Consideramos a la fecha como index

data_2.index=data_2['Date']
data_2
```

```python
# Desagregamos la columna los casos diarios

data_3=data_2.Argentina.diff()
data_3
```
