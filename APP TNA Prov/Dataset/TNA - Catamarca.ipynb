# Tratamiento Transferencias no Automáticas - Catamarca

```python
import pandas as pd
import seaborn
import matplotlib as plt
import numpy as np
import plotly.plotly as py
import plotly.figure_factory as ff
import datetime
```

```python
tna = pd.read_excel('TNA Agregado.xlsx',
                   parse_dates=['mes'])
```

```python
tna_catam = tna[tna['ubicacion_geografica_desc'] == 'Provincia de Catamarca']
tna_catam
```

```python
tna_catam = tna_catam.replace(['Transf. a Inst. Prov. y Mun. para Fin. Gastos Corrientes','Transf. a Inst. Prov. y Mun. para Financ. Gastos de Capital'], ['TNA Corrientes','TNA de Capital'])
```

*Evolución de las transferencias corrientes y de capital en términos nominales*

```python
tna_1=pd.DataFrame(tna_catam, columns=['ejercicio_presupuestario', 'principal_desc', 'credito_devengado'])
```

```python
# Año 2015
tna_2015 = pd.DataFrame(tna_1[tna_1['ejercicio_presupuestario'] == 2015], columns=['principal_desc', 'credito_devengado'])
tna_2015 = pd.DataFrame(tna_2015.groupby('principal_desc').sum().reset_index())
tna_2015.rename(index=str, columns={'credito_devengado':'2015'}, inplace=True)

# Año 2016
tna_2016 = pd.DataFrame(tna_1[tna_1['ejercicio_presupuestario'] == 2016], columns=['principal_desc', 'credito_devengado'])
tna_2016 = pd.DataFrame(tna_2016.groupby('principal_desc').sum().reset_index())
tna_2016.rename(index=str, columns={'credito_devengado':'2016'}, inplace=True)

tabla = pd.merge(tna_2015, tna_2016, how='left', on='principal_desc')

# Año 2017
tna_2017 = pd.DataFrame(tna_1[tna_1['ejercicio_presupuestario'] == 2017], columns=['principal_desc', 'credito_devengado'])
tna_2017 = pd.DataFrame(tna_2017.groupby('principal_desc').sum().reset_index())
tna_2017.rename(index=str, columns={'credito_devengado':'2017'}, inplace=True)

# Año 2018
tna_2018 = pd.DataFrame(tna_1[tna_1['ejercicio_presupuestario'] == 2018], columns=['principal_desc', 'credito_devengado'])
tna_2018 = pd.DataFrame(tna_2018.groupby('principal_desc').sum().reset_index())
tna_2018.rename(index=str, columns={'credito_devengado':'2018'}, inplace=True)

tabla_1 = pd.merge(tna_2017, tna_2018, how='left', on='principal_desc')

# Año 2019
tna_2019 = pd.DataFrame(tna_1[tna_1['ejercicio_presupuestario'] == 2019], columns=['principal_desc', 'credito_devengado'])
tna_2019 = pd.DataFrame(tna_2019.groupby('principal_desc').sum().reset_index())
tna_2019.rename(index=str, columns={'credito_devengado':'2019'}, inplace=True)

tablas = pd.merge(tabla, tabla_1, how='left', on='principal_desc')
tablas_1 = pd.merge(tablas, tna_2019, how='left', on='principal_desc')
print('Transferencias a la Provincia de Catamarca en el Primer Cuatrimestre de cada año (en millones de $)')
tablas_1
```

```python
tablas_1['2015vs2016'] = tablas_1['2016'] / tablas_1['2015'] - 1
tablas_1['2016vs2017'] = tablas_1['2017'] / tablas_1['2016'] - 1
tablas_1['2017vs2018'] = tablas_1['2018'] / tablas_1['2017'] - 1
tablas_1['2018vs2019'] = tablas_1['2019'] / tablas_1['2018'] - 1
print('Transferencias a la Provincia de Catamarca en el Primer Cuatrimestre de cada año. Variación nominal.')
pd.DataFrame(tablas_1, columns=['principal_desc','2015vs2016', '2016vs2017', '2017vs2018', '2018vs2019'])
```

*Evolución de las transferencias corrientes y de capital en términos reales*

```python
tna_2=pd.DataFrame(tna_catam, columns=['ejercicio_presupuestario', 'principal_desc', 'Devengado Real'])
```

```python
# Año 2015
tna_2015_r = pd.DataFrame(tna_2[tna_2['ejercicio_presupuestario'] == 2015], columns=['principal_desc', 'Devengado Real'])
tna_2015_r = pd.DataFrame(tna_2015_r.groupby('principal_desc').sum().reset_index())
tna_2015_r.rename(index=str, columns={'Devengado Real':'2015'}, inplace=True)

# Año 2016
tna_2016_r = pd.DataFrame(tna_2[tna_2['ejercicio_presupuestario'] == 2016], columns=['principal_desc', 'Devengado Real'])
tna_2016_r = pd.DataFrame(tna_2016_r.groupby('principal_desc').sum().reset_index())
tna_2016_r.rename(index=str, columns={'Devengado Real':'2016'}, inplace=True)

tabla_r = pd.merge(tna_2015_r, tna_2016_r, how='left', on='principal_desc')

# Año 2017
tna_2017_r = pd.DataFrame(tna_2[tna_2['ejercicio_presupuestario'] == 2017], columns=['principal_desc', 'Devengado Real'])
tna_2017_r = pd.DataFrame(tna_2017_r.groupby('principal_desc').sum().reset_index())
tna_2017_r.rename(index=str, columns={'Devengado Real':'2017'}, inplace=True)

# Año 2018
tna_2018_r = pd.DataFrame(tna_2[tna_2['ejercicio_presupuestario'] == 2018], columns=['principal_desc', 'Devengado Real'])
tna_2018_r = pd.DataFrame(tna_2018_r.groupby('principal_desc').sum().reset_index())
tna_2018_r.rename(index=str, columns={'Devengado Real':'2018'}, inplace=True)

tabla_1_r = pd.merge(tna_2017_r, tna_2018_r, how='left', on='principal_desc')

# Año 2019
tna_2019_r = pd.DataFrame(tna_2[tna_2['ejercicio_presupuestario'] == 2019], columns=['principal_desc', 'Devengado Real'])
tna_2019_r = pd.DataFrame(tna_2019_r.groupby('principal_desc').sum().reset_index())
tna_2019_r.rename(index=str, columns={'Devengado Real':'2019'}, inplace=True)

tablas_r = pd.merge(tabla_r, tabla_1_r, how='left', on='principal_desc')
tablas_1_r = pd.merge(tablas_r, tna_2019_r, how='left', on='principal_desc')
```

```python
tablas_1_r['2015vs2016'] = tablas_1_r['2016'] / tablas_1_r['2015'] - 1
tablas_1_r['2016vs2017'] = tablas_1_r['2017'] / tablas_1_r['2016'] - 1
tablas_1_r['2017vs2018'] = tablas_1_r['2018'] / tablas_1_r['2017'] - 1
tablas_1_r['2018vs2019'] = tablas_1_r['2019'] / tablas_1_r['2018'] - 1
tablas_1_r['2015vs2019'] = tablas_1_r['2019'] / tablas_1_r['2015'] - 1
print('Transferencias a la Provincia de Catamarca en el Primer Cuatrimestre de cada año. Variación en Términos Reales.')
pd.DataFrame(tablas_1_r, columns=['principal_desc','2015vs2016', '2016vs2017', '2017vs2018', '2018vs2019', '2015vs2019'])
```
